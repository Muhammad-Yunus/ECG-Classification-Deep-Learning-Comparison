### Inference Model Denoising & Classification


```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras import models
from keras import backend as K
```

- load model

```python
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true))) 
```

```python
denoising_Conv_AE_model = models.load_model('denoising_conv_AE.h5', custom_objects={'rmse': rmse})
```

```python
cnn_model = models.load_model('CNN_Classification_model.h5')
dnn_model = models.load_model('DNN_Classification_model.h5')
rnn_model = models.load_model('RNN_Classification_model.h5')
```

```python
label_names = ['AF', 'N']
fs = 250
```

```python
dataset_folder ="dataset/"
filenames = []
for filename in os.listdir(dataset_folder):
    if filename.find("AFDB_pasien_") > -1:
        filenames.append(filename)
```

```python
filenames
```

- load inference data

```python
# scale data from 0 to 1
def scaler(X):
    res = []
    for x in X :
        idx = np.max(np.nonzero(x))
        x[idx+1:] = x.min()
        res.append((x - x.min())/(x.max() - x.min()))
    return np.array(res)
```

```python
inference_data = {}
for name in filenames:
    infer = pd.read_csv(dataset_folder + name, header=None)
    inference_data[name] = infer
```

```python
inference_data[filenames[0]].head()
```

```python
mat_infer_data = {}
for name in filenames:
    infer_df = inference_data[name]
    X = infer_df.values
    X = scaler(X)
    mat_infer_data[name] = X
    
```

- Apply denoising model to inference data

```python
Conv_AE_den_res = {}
for name in filenames:
    Conv_AE_den_res[name] = denoising_Conv_AE_model.predict(mat_infer_data[name].reshape(len(mat_infer_data[name]), 300, 1))
    print("denoising data - %s" % name)
```

```python
def plot_decoded_signal(decoded_signal, original_signal, name):
    time = np.arange(len(decoded_signal)) / fs
    plt.figure(figsize=(15,4))

    plt.subplot(1,2,1)
    plt.plot(decoded_signal)
    plt.title("Denoised Signal - " + name)
    plt.legend(['channel 1', 'channel 2'])
    plt.grid()

    plt.subplot(1,2,2)
    plt.plot(original_signal)
    plt.title("Original Signal")
    plt.legend(['channel 1', 'channel 2'])
    plt.grid()

    plt.tight_layout()
    plt.show()
```

- Check result

```python
name = filenames[0]
idx = 130

decoded_signal = Conv_AE_den_res[name][idx]
signal = mat_infer_data[name][idx]

plot_decoded_signal(decoded_signal, signal, name="Conv-AE")
```

- Reshape matrix result of denoising

```python
Conv_AE_mat_3d = {}
for name in filenames:
    Conv_AE_mat_3d[name] = Conv_AE_den_res[name].reshape(len(Conv_AE_den_res[name]), 300, 1)
```

- Apply to CNN, DNN & RNN model

```python
cnn_classif_result = {}
dnn_classif_result = {}
rnn_classif_result = {}
for name in filenames:
    cnn_classif_result[name] = cnn_model.predict(Conv_AE_mat_3d[name])
    dnn_classif_result[name] = dnn_model.predict(Conv_AE_mat_3d[name].reshape(len(Conv_AE_den_res[name]), 300))
    rnn_classif_result[name] = rnn_model.predict(Conv_AE_mat_3d[name])
    print("predicting data - %s" % name)
```

- Check result

```python
def plot_result(data, label='Deep-AE', confidence=0.0):
    confidence = confidence*100
    # show results
    time = np.arange(len(data)) / fs
    
    plt.figure(figsize=(6,4))
    plt.title("Classification Result : %s - confidence (%.2f%%)" % (label, confidence))
    plt.xlabel("samples - n")
    plt.plot(time, data,"b", label="channel 1")
    plt.tight_layout()
    plt.legend()
    plt.grid()
    plt.show()
```

- parameter

```python
curr_model = 'cnn' # change to 'cnn' / 'dnn' / 'rnn'
```

```python
pred_result = []
for name in filenames :
    pred_res = cnn_classif_result[name] if curr_model == 'cnn' else dnn_classif_result[name] if curr_model == 'dnn' else rnn_classif_result[name]
    print("\n\n\nPrediction Result - %s\n" % name)
    for i, res in enumerate(pred_res):
        predicted_id = np.argmax(res)
        label = label_names[predicted_id]
        confidence = res[predicted_id]
        label = label if confidence > 0.50 else '?' # threshold for confidence level , set to 70%
        pred_result.append([name, i, label, confidence])
        
        print('%s \t index - %d \t %s \t %.2f %%' % (name, i, label, confidence*100))
```

```python
res = pred_result[90]
data = Conv_AE_den_res[name][res[1]]
plot_result(data, res[2], res[3])
```

```python
res = pred_result[-19]
data = Conv_AE_den_res[name][res[1]]
plot_result(data, res[2], res[3])
```

```python
pred_result_df = pd.DataFrame(pred_result, columns=['filename', 'idx', 'label', 'confidence'])
```

```python
pred_result_df.head()
```

```python
for name in filenames:
    print(name)
    count = dict(pred_result_df[pred_result_df['filename'] == name]['label'].value_counts())
    for key in count :
        print("%s \t: %d" % (key, count[key]))
    print('\n')
```

```python
def plot_donat(porsion, title, labels, colors=['red', 'green', 'blue']):
    plt.figure(figsize=(10,5))

    circle = plt.Circle( (0,0), 0.7, color='white')
    plt.pie(porsion, 
            labels=labels, 
            colors=colors,
            autopct='%1.1f%%')

    plt.title(title)
    p = plt.gcf()
    p.gca().add_artist(circle)
    plt.show()
```

```python
for name in filenames:
    count = dict(pred_result_df[pred_result_df['filename'] == name]['label'].value_counts())
    plot_donat(count.values(), name, count.keys())
```

- save result as csv

```python
pred_result_df.to_csv(dataset_folder + 'inference_result_%s_Conv_AE_pasien_AF_dan_Normal.csv' % curr_model)
```
